#### Import libraries & read in UKF output files

```python
import numpy as np
import pandas as df
import matplotlib.pyplot as plt
from glob import glob
```

```python
data_path = '.\\output\\output*'
files = glob(data_path)
dataset_1 = df.read_csv(files[0])
dataset_2 = df.read_csv(files[1])
```

#### Analyze the NIS values

```python
df3_095 = 7.815
df2_095 = 5.991
df3_005 = 0.352
df2_005 = 0.102

radar_max = [df3_095 for i in range(0,500)]
laser_max = [df2_095 for i in range(0,500)]
radar_min = [df3_005 for i in range(0,500)]
laser_min = [df2_005 for i in range(0,500)]

plt.figure(figsize = [15,5])
plt.subplot(1,2,1)
plt.plot(dataset_1.loc[dataset_1['sensor'] == 'R', ['NIS']], 'g', label = "NIS Values")
plt.plot(range(500), radar_max, '--k', label = "7.815")
plt.plot(range(500), radar_min, '--k', label = "0.352")
plt.title("RADAR NIS Values")
plt.xlabel('Time Step')
plt.ylabel('NIS Value')
plt.legend()
plt.ylim([0,16])
plt.subplot(1,2,2)
plt.plot(dataset_1.loc[dataset_1['sensor'] == 'L', ['NIS']], label = "NIS Values")
plt.plot(range(500), laser_max, '--k', label = "5.991")
plt.plot(range(500), laser_min, '--k', label = "0.102")
plt.title("LIDAR NIS Values")
plt.xlabel('Time Step')
plt.ylabel('NIS Value')
plt.ylim([0,16])
plt.legend()
plt.show()
```

```python
radar_NIS = dataset_1.loc[(dataset_1['sensor'] == 'R'), ['NIS']]
lidar_NIS = dataset_1.loc[dataset_1['sensor'] == 'L', ['NIS']]
radar_NIS_percentage = len(radar_NIS[(radar_NIS['NIS'] < df3_095) & (radar_NIS['NIS'] > df3_005)])/len(radar_NIS)*100
lidar_NIS_percentage = len(lidar_NIS[(lidar_NIS['NIS'] < df2_095) & (lidar_NIS['NIS'] > df2_005)])/len(lidar_NIS)*100
print("Percentage of RADAR NIS points between 5% and 95% threshold: {:.2f}%".format(radar_NIS_percentage))
print("Percentage of LIDAR NIS points between 5% and 95% threshold: {:.2f}%".format(lidar_NIS_percentage))
```

#### Visualize the accuracy of the UKF

```python
plt.figure(figsize = [15,10])
plt.subplot(2,2,1)
plt.plot(dataset_1['x_gt'], dataset_1['y_gt'], 'k', label="Ground Truth")
plt.plot(dataset_1['px'], dataset_1['py'], '--r', label = "Estimations")
plt.title("Estimated vs Ground truth - Position")
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.subplot(2,2,2)
plt.plot(range(len(dataset_1)), (dataset_1['vy_gt']**2 + dataset_1['vx_gt']**2)**0.5, 'k', label="Ground Truth")
plt.plot(range(len(dataset_1)), (dataset_1['vx']**2 + dataset_1['vy']**2)**0.5, '--r', label = "Estimations")
plt.title("Estimated vs Ground truth - Absolute Velocity")
plt.xlabel('Time Step')
plt.ylabel('Velocity [m/s]')
plt.legend()
plt.subplot(2,2,3)
plt.plot(range(len(dataset_1)), dataset_1['yaw_gt'], 'k', label="Ground Truth")
plt.plot(range(len(dataset_1)), dataset_1['yaw'], '--r', label = "Estimations")
plt.title("Estimated vs Ground truth - Yaw Angle")
plt.xlabel('Time Step')
plt.ylabel('Yaw Angle [rad]')
plt.legend(loc="upper left")
plt.subplot(2,2,4)
plt.plot(range(len(dataset_1)), dataset_1['yawdot_gt'], 'k', label="Ground Truth")
plt.plot(range(len(dataset_1)), dataset_1['yawdot'], '--r', label = "Estimations")
plt.title("Estimated vs Ground truth - Yaw Rate")
plt.xlabel('Time Step')
plt.ylabel('Yaw Rate [rad/s2]')
plt.legend()
plt.show()
```
